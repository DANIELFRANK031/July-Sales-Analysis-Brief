```python
##Extracting the following attributes into a new dataset called  newdf.
newdf=df[['OrderedAtDate ','Product','CategoryGroup','Category','Quantity','Unit Price','shippingStatus','DeliveryDate','CustomerType','OrderChannel']]
```

```python
##checking for the extracted data set
newdf.head()
```

```python
##exploring newdata set to show information after attributes extractions
newdf.info()
```

```python
##Converting Orderdate and DeliveveryDate from objects to Datetime format
newdf['OrderedAtDate ']=pd.to_datetime(newdf['OrderedAtDate '])
newdf['DeliveryDate']=pd.to_datetime(newdf['DeliveryDate'])
```

```python
 newdf.head()
```

```python
##Changes successfully appended
newdf.info()
```

```python
##Converting Unit Price from str to float for statistical analysis
newdf['Unit Price'] = newdf['Unit Price'].str.replace(',','').astype(float)
```

```python
newdf.info()
```

```python
##Checking for null values.. 
newdf.isnull()
```

```python
##Converting nan values on DeliveryDate to not specified
newdf['DeliveryDate'] = df['DeliveryDate'].astype(str).replace('nan', 'Not specified')
```
